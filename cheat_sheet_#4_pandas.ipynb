–ò–º–µ–µ—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –∏ –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –¥–ª—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö. –ë–æ–ª—å—à–µ –¥–µ—Ç–∞–ª–µ–π –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ –≤ [**–¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏.**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html)

### .head()

### .tail()

–í—ã–≤–æ–¥–∏—Ç **–ø–æ—Å–ª–µ–¥–Ω–∏–µ** –∑–∞–ø–∏—Å–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞. –í—ã–≤–µ–¥–µ—Ç —Å—Ç–æ–ª—å–∫–æ –∑–∞–ø–∏—Å–µ–π, —Å–∫–æ–ª—å–∫–æ —É–∫–∞–∂–µ–º –≤ —Å–∫–æ–±–∫–∞—Ö –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –≤—ã–≤–æ–¥–∏—Ç 5 —Å—Ç—Ä–æ–∫.



````python

df.tail() # –≤—ã–≤–µ–¥–µ—Ç 5 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ–∫
df.tail(10) # –≤—ã–≤–µ–¥–µ—Ç 10 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ—á–µ–∫


````


üí° –° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–æ–≤ `.head()` –∏ .`tail()` –º—ã –º–æ–∂–µ–º –Ω–µ —Ç–æ–ª—å–∫–æ —Å–º–æ—Ç—Ä–µ—Ç—å –¥–∞–Ω–Ω—ã–µ, –Ω–æ –∏ —Å–æ–∑–¥–∞–≤–∞—Ç—å —Å—Ä–µ–∑—ã –∏–∑ –ø–µ—Ä–≤—ã—Ö –∏–ª–∏ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ–∫ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –Ω–∞–º –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å—Ä–µ–∑ –∏–∑ –¢–û–ü-5 –Ω–∞–∏–±–æ–ª–µ–µ –∏ –Ω–∞–∏–º–µ–Ω–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –≤–∏–¥–æ–≤ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ–∫, –º–æ–∂–Ω–æ –≤–∑—è—Ç—å —Å—Ä–µ–∑ –æ–¥–Ω–∏–º –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –∏ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.



### .info()

–í—ã–≤–æ–¥–∏—Ç –æ–±—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å—Ç—Ä–æ–∫–∞—Ö –∏ –∫–æ–ª–æ–Ω–∫–∞—Ö: 

- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –∏ –∫–æ–ª–æ–Ω–æ–∫
- –∑–∞–ø–æ–ª–Ω—è–µ–º–æ—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏–π
- —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö



````python

df.info()


````

### .describe()

–í—ã–≤–æ–¥–∏—Ç –æ—Å–Ω–æ–≤–Ω—ã–µ –º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —á–∏—Å–ª–æ–≤—ã–º –∏ —Ç–µ–∫—Å—Ç–æ–≤—ã–º –ø–æ–ª—è–º:

- **`count`** ‚Äî –ø–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
- **`mean`** ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
- **`min`** ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
- **`max`** ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
- **`std`** ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
- **`50%`** ‚Äî –º–µ–¥–∏–∞–Ω–∞
- **`25%`** ‚Äî 25-–π –∫–≤–∞–Ω—Ç–∏–ª—å
- **`75%`** ‚Äî 75-–π –∫–≤–∞–Ω—Ç–∏–ª—å

–ß—Ç–æ–±—ã —É–∫–∞–∑–∞—Ç—å, –æ–ø–∏—Å–∞–Ω–∏–µ –∫–∞–∫–∏—Ö –ø–æ–ª–µ–π –Ω—É–∂–Ω–æ –≤—ã–≤–æ–¥–∏—Ç—å, –º–æ–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å –∞—Ä–≥—É–º–µ–Ω—Ç¬†`include`, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥:

- –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –≤—ã–≤–æ–¥–∏—Ç –æ–ø–∏—Å–∞–Ω–∏—è —á–∏—Å–ª–æ–≤—ã—Ö –ø–æ–ª–µ–π, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å)
- —Å—Ç—Ä–æ–∫—É¬†`'all'`, –µ—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º –≤—ã–≤–µ—Å—Ç–∏ –≤—Å–µ –ø–æ–ª—è



````python

df.describe() # –≤—ã–≤–µ–¥–µ—Ç –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º —Å —á–∏—Å–ª–æ–≤—ã–º–∏ —Ç–∏–ø–∞–º–∏
df['column'].describe() # –≤—ã–≤–µ–¥–µ—Ç –¥–∞–Ω–Ω—ã–µ –¥–ª—è –æ–¥–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞

df.describe(include = 'all') # –≤—ã–≤–µ–¥–µ—Ç –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
df.describe(include = 'object') # –≤—ã–≤–µ–¥–µ—Ç –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ —Å –¥–∞–Ω–Ω—ã–º–∏ —Ç–∏–ø object


````


üí° –ù–µ –∑–∞–±—ã–≤–∞–π, —á—Ç–æ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —ç—Ç–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –≤ —Å–µ—Ä–µ–¥–∏–Ω–µ —è—á–µ–π–∫–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ –æ—Ç–æ–±—Ä–∞–∑—è—Ç—Å—è.


## –û–±—Ä–∞—â–µ–Ω–∏–µ –∫ —Å—Ç–æ–ª–±—Ü–∞–º –∏ —Å—Ç—Ä–æ–∫–∞–º

### df[ 'column_name' ]

–ù–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–æ–≤–æ–º —Ñ–æ—Ä–º–∞—Ç–µ –≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö —Å–∫–æ–±–∫–∞—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ **—Å–ø–∏—Å–æ–∫** –Ω–∞–∑–≤–∞–Ω–∏–π –¥–ª—è –≤—ã–≤–æ–¥–∞ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤.



````python

df['–∏–º—è —Å—Ç–æ–ª–±—Ü–∞'] # –∏–º—è –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±—ã–º
df[['–∏–º—è —Å—Ç–æ–ª–±—Ü–∞ 1', '–∏–º—è —Å—Ç–æ–ª–±—Ü–∞ 2']] # –¥–≤–æ–π–Ω—ã–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ —Å–∫–æ–±–∫–∏ –¥–ª—è —Å–ø–∏—Å–∫–∞


````

### .set_index( )

–ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞ —Å—Ç—Ä–æ–∫–∏ —Å –ø–æ–º–æ—â—å—é –æ–¥–Ω–æ–≥–æ –∏–∑ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞. 

–ê—Ä–≥—É–º–µ–Ω—Ç `inplace` –≤–ª–∏—è–µ—Ç –Ω–∞ —Ç–æ, –∏–∑–º–µ–Ω—è–µ–º –ª–∏ –º—ã –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º ‚Äì `True,` –∏–ª–∏ –Ω–µ—Ç ‚Äì `False`



````python
df.set_index('user_id', inplace = True) 
# –Ω–∞–∑–Ω–∞—á–∞–µ–º –∫–æ–ª–æ–Ω–∫—É user_id –∏–Ω–¥–µ–∫—Å–æ–º –∏ –ø–µ—Ä–µ–∑–∞–ø–∏—Å—ã–≤–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç


````

### .reset_index( )

–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∏–Ω–¥–µ–∫—Å —Å—Ç–∞–ª —Å—Ç–æ–ª–±—Ü–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞. –ü—Ä–∏–º–µ–Ω—è–µ–º –≤ —Å–ª—É—á–∞—è—Ö:

- –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –≤–µ—Ä–Ω—É—Ç—å –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å—Ç–æ–ª–±–µ—Ü, –∫–æ—Ç–æ—Ä—ã–π –º—ã —Ä–∞–Ω–µ–µ –Ω–∞–∑–Ω–∞—á–∏–ª–∏ –∫–∞–∫ —Å—Ç–æ–ª–±–µ—Ü –∏–Ω–¥–µ–∫—Å–∞, –Ω–∞–ø—Ä–∏–º–µ—Ä user_id
- –µ—Å–ª–∏ –º—ã –≤—ã–∑–≤–∞–ª–∏ –≤—Å–µ–≥–æ –æ–¥–∏–Ω —Å—Ç–æ–ª–±–µ—Ü –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –ø–æ–ª—É—á–∏–ª–∏ –æ–¥–Ω–æ–º–µ—Ä–Ω—É—é —Ç–∞–±–ª–∏—Ü—É ‚Äî `series`, –∞ —Ç–µ–ø–µ—Ä—å —Ö–æ—Ç–∏–º, —á—Ç–æ–±—ã `series` —Å—Ç–∞–ª–∞ `dataframe`

–ê—Ä–≥—É–º–µ–Ω—Ç `inplace` –≤–ª–∏—è–µ—Ç –Ω–∞ —Ç–æ, –∏–∑–º–µ–Ω—è–µ–º –ª–∏ –º—ã –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º ‚Äì `True`, –∏–ª–∏ –Ω–µ—Ç ‚Äì `False` (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é `False`)



````python

df.reset_index(inplace = True) # —Ç–µ–ø–µ—Ä—å —Å—Ç–∞—Ä—ã–µ –∏–Ω–¥–µ–∫—Å—ã –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ —Å—Ç–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Å—Ç–æ–ª–±—Ü–∞


````


üí° –ï—Å–ª–∏ –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ `.reset_index()` –º—ã –¥–æ —ç—Ç–æ–≥–æ –Ω–∞–∑–Ω–∞—á–∞–ª–∏ –∫–∞–∫–æ–π-—Ç–æ —Å—Ç–æ–ª–±–µ—Ü –∏–Ω–¥–µ–∫—Å–æ–º, –æ–Ω –ø—Ä–æ—Å—Ç–æ –≤–µ—Ä–Ω–µ—Ç—Å—è –æ–±—Ä–∞—Ç–Ω–æ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.

–ï—Å–ª–∏ –∏–Ω–¥–µ–∫—Å –±—ã–ª –æ—á–µ—Ä–µ–¥–Ω–æ—Å—Ç—å—é –æ—Ç 0 –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å—Ç—Ä–æ–∫, —Ç–æ —Ç–µ–ø–µ—Ä—å –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –µ—Å—Ç—å —Å—Ç–æ–ª–±–µ—Ü —Å –Ω—É–º–µ—Ä–∞—Ü–∏–µ–π.



### .loc[ ]

–î–æ—Å—Ç–∞–µ—Ç —Å—Ä–µ–∑—ã –∏ —Ñ–∏–ª—å—Ç—Ä—É–µ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º

1. –º–æ–∂–Ω–æ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—Ç—å —Å—Ä–µ–∑ —Å—Ç—Ä–æ–∫ —Å–æ –≤—Å–µ–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏
2. –º–æ–∂–Ω–æ –≤—ã—Ç–∞—Å–∫–∏–≤–∞—Ç—å —Å—Ä–µ–∑ –∏–∑ —Å—Ç—Ä–æ–∫ –∏ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫, —Ç–æ–≥–¥–∞ –ø–µ—Ä–µ–¥–∞–µ–º –≤—Ç–æ—Ä—ã–º —Å–ø–∏—Å–∫–æ–º –∫–æ–ª–æ–Ω–∫–∏
3. –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö

–ü—Ä–∏–Ω–∏–º–∞–µ—Ç –≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ–ª–æ–Ω–æ–∫ –∏ —Å—Ç—Ä–æ–∫ –≤ –≤–∏–¥–µ —Å–ø–∏—Å–∫–∞.

### .drop()

–ê—Ä–≥—É–º–µ–Ω—Ç `inplace` –≤–ª–∏—è–µ—Ç –Ω–∞ —Ç–æ, –∏–∑–º–µ–Ω—è–µ–º –ª–∏ –º—ã –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º ‚Äì `True`, –∏–ª–∏ –Ω–µ—Ç ‚Äì `False`(–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é).

–í–æ–∫—Ä—É–≥ `drop` –º–Ω–æ–∂–µ—Å—Ç–≤–æ –º–µ—Ç–æ–¥–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∑–≤–æ–ª—è—é—Ç —É–¥–∞–ª—è—Ç—å –Ω–µ —Ç–æ–ª—å–∫–æ —Ü–µ–ª—ã–π —Å—Ç–æ–ª–±–µ—Ü, –Ω–æ –∏ —Ä–∞–∑–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –ø–æ —É—Å–ª–æ–≤–∏—è–º.


````python

new_df = df.drop(columns = 'name') 
# —É–¥–∞–ª–∏–º —Å—Ç–æ–ª–±–µ—Ü name –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –ø–æ–ª–æ–∂–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –Ω–æ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é 
# –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –Ω–µ –∏–∑–º–µ–Ω–∏—Ç—Å—è

df.drop(columns = 'name', inplace = True) # —É–¥–∞–ª–∏–º –∫–æ–ª–æ–Ω–∫—É name –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ df –∏ –ø–µ—Ä–µ–∑–∞–ø–∏—à–µ–º –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º

df.dropna() # —É–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ –ø—É—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º

df.drop_duplicates() # —É–¥–∞–ª–∏–º –¥—É–±–ª–∏—Ä—É—é—â–∏–µ—Å—è —Å—Ç—Ä–æ–∫–∏


````


### **sort_values( )**

–û—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º.

- –æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç `by` ‚Äì —É–∫–∞–∑—ã–≤–∞–µ—Ç, –ø–æ –∫–∞–∫–æ–π –∫–æ–ª–æ–Ω–∫–µ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å
- —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ–Ω–∏–º–∞–µ—Ç –≤—Å–µ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö
- –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é: —á—Ç–æ–±—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ —É–±—ã–≤–∞–Ω–∏—é, –Ω—É–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞: `ascending = False`


üí° –ê –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø–æ –∏–Ω–¥–µ–∫—Å–∞–º –µ—Å—Ç—å –º–µ—Ç–æ–¥  `.sort_index()`



````python

df.sort_values(by='column') # –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ—Ç –¥–∞—Ç–∞—Å–µ—Ç df –ø–æ –∫–æ–ª–æ–Ω–∫–µ column –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é

df.sort_values(by='column', ascending=False) # –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ—Ç –¥–∞—Ç–∞—Å–µ—Ç df –ø–æ –∫–æ–ª–æ–Ω–∫–µ column –ø–æ —É–±—ã–≤–∞–Ω–∏—é

df.sort_index() # –≤—ã—Å—Ç—Ä–æ–∏—Ç –∏–Ω–¥–µ–∫—Å—ã –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é


````
